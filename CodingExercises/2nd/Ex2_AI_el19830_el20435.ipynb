Σε αυτό το σημείο θα γίνει μια αναβάθμιση του συστήματος έτσι ώστε να παράγονται καλύτερες  συστάσεις οι όποιες θα λαμβάνουν υπόψιν και τις προτιμήσεις του χρήστη, οι όποιες θα εξάγονται από τις αξιολογήσεις που έχει κάνει μέχρι στιγμής. 

Η εκπαίδευση του recommender θα γίνεται ως εξής:

Για κάθε ταινία θα υπάρχει ένα score το όποιο αρχικά θε είναι ίσο με 0 και θα διαμορφώνεται από τις αξιολογήσεις κάθε user. Έτσι για έναν χρήστη με βάση τις αξιολογήσεις που υπάρχουν στο αρχείο train_ratings θα πρέπει:

1.	Για κάθε ταινία που έχει βαθμολογήσει να βρίσκονται οι κοινές ταινίες ανά κλίμακα και στο μέχρι τώρα σκορ κάθε παρόμοιας ταινίας θα προστίθεται ένα βάρος το όποιο θα μπορούσε να είναι το ποσοστό ομοιότητας της ταινίας (δηλαδή ένα βάρος για κάθε κλίμακα, αν δύο ταινίες μοιάζουν ενισχύουμε το βάρος που προσθέτουμε από το να μοιάζουν λιγότερο) επί τον βαθμό που έχει βάλει ο χρήστης για την αρχική ταινία (διαφορετικό είναι ο χρήστης να έχει βάλει 5/5 ή 1/5 σε μια ταινία από 3/5). 

2. Στη συνέχεια, ανάλογα με το σκορ που έχει σχηματιστεί για κάθε ταινία, θα επιλέγεται αν αυτή θα μπορούσε να είναι προτεινόμενη για τον χρήστη ή όχι και θα μετράμε το πόσο καλά τα πήγε το σύστήμά μας με βάση ορισμένες μετρικές.

Η λογική πίσω από την παραπάνω διαδικασία είναι ότι παρόμοιες ταινίες θα έχουν ανάλογο βαθμό. Για παράδειγμα αν ένας χρήστης έχει αξιολογήσει αρκετές  ταινίες οι όποιες είναι sci-fiction με 5/5 τότε μια ταίνια sci-fiction την όποια δεν έχει δεί λογικά θα του αρέσει και θα έπρεπε να την προτείνουμε.

Μετά την εκπαίδευση του συστήματος σας καλείστε να δοκιμάσετε τον recommender που κατασκευάσατε στην πράξη. Για τον σκοπό αυτό θα φορτώσετε το αρχείο test_ratings.csv όπου περιέχονται οι αξιολογήσεις του ίδιου χρήστη για άλλες ταινίες. Το σύστημά σας πρέπει να προβλέπει αν μια ταινία θα πρέπει να προταθεί στον χρήστη. Μια ταίνια θεωρούμε οτι προτείνεται στον  χρήστη, αν έχει βαθμό μεγαλύτερο του 3 στο test_ratings. Συνεπώς για την επίβλεψη του συστήματός σας θα πρέπει για κάθε μια από τις ταινίες του αρχείου test_ratings.csv να προβλέψετε εάν θα αρέσει στο χρήστη ή όχι ώστε να την προτείνετε. 

ΠΡΟΣΟΧΗ! Η "βαθμολογία" που θα υπολογίσετε για κάθε ταινία κατά το training δεν είναι απαραίτητα πρόβλεψη της βαθμολογίας που θα έβαζε ο χρήστης. 

Στην συνέχεια, σε συνδιασμό με τις πραγματικές απαντήσεις του χρήστη θα αξιολογήσετε το σύστημά σας χρησιμοποιόντας τις μετρικές: precision, recall, f1 οι οποίες είναι οι πλέον γνωστές μετρικές και ευρέως χρησιμοποιούμενες τεχνικές για την επίβλεψη-μέτρηση απόδοσης ανάλογων συστημάτων.

1. Precision: Δείχνει πόσο ακριβές είναι το σύστημα. Υπολογίζει πόσα από τα στιγμιότυπα τα όποια προβλέψαμε ότι ανήκουν σε μια κλάση όντως ανήκουν σε αυτή. Η μετρική αυτή μας δίνει μια εικόνα σχετικά με τον αριθμό των ταινίων που προβλέψαμε ως προτεινόμενες ενώ δεν θα έπρεπε.

2. Recall: Υπολογίζει πόσα από τα στιγμιότυπα που ανήκουν σε μια κλάση (π.χ. προτεινόμενες ταινίες) προβλέφθηκαν σωστά.

3. F1: Είναι ένας μέσος μεταξύ των παραπάνω δυο μετρικών, έτσι ώστε να διατηρείται μια ισορροπία μεταξύ τους. Υπολογίζεται από την παρακάτω σχέση:

$$F_1=2\frac{Precision\times{Recall}}{Precision+Recall}$$

Οι παραπάνω συναρτήσεις παρέχονται από την βιβλιοθήκη scikit-learn.

````python
def clean_text(text):
  text = text.replace(u'\xa0', u'')
  text = text.replace(u"'", u'')
  return text
````

# ΜΕΡΟΣ 1
## 1.α. Ορισμός Κατηγορημάτων

Θα κατασκευάσουμε μια βάση γνώσης για την Prolog χρησιμοποιώντας το αρχείο `movies_metadata.csv`.

Για κάθε record ταινίας, έχουμε τα εξής:
-  Απλοποιούμε τα string attributes (π.χ. αφαίρεση punctuation) για ευκολότερη διαχείριση τους και καλύτερη γενίκευση του μοντέλου, χρησιμοποιώντας τη συνάρτηση `clean_text()`.
-  Για κάθε attribute ορίζουμε ένα κατηγόρημα που λαμβάνει δύο ορίσματα, όπου το πρώτο είναι ο τίτλος της τρέχουσας ταινίας και το δεύτερο είναι η τιμή του συγκεκριμένου attribute. Για παράδειγμα, ορίζουμε το κατηγόρημα `budget(<movie_title>, <movie_budget>)`, το οποίο δηλώνει ότι η ταινία με τίτλο `movie_title` έχει budget `movie_budget`. Ανάλογα με τη τιμή κάθε attribute στο csv αρχείο, ορίζουμε κατάλληλες δομές δεδομένων για την επεξεργασία του, όπως για παράδιειγμα λίστες και λεξικά.


Ακόμα, εισάγουμε τα κατηγορήματα σε μία αρχικά κενή λίστα ονόματι `literals` την οποία στη συνέχεια διατρέχουμε ώστε να γεμίσουμε τη βάση δεδομένων της prolog.

````python
def handle_similarity(s, q, i, show = True):
    for soln in q:
      m = soln['M']
      if m not in s:
        s.add(soln['M'])
    q.close()
    if show:
        print("Similarity of " + str(i) + ": ")
        print(list(s))
        print()
    return s

def find_similar(movie):
  s1 = set()
  q1 = prolog.query("find_similar_1('" + movie +"', M)")

  s2 = set()
  q2 = prolog.query("find_similar_2('" + movie +"', M)")

  s3 = set() 
  q3 = prolog.query("find_similar_3('" + movie +"', M)")

  s4 = set()
  q4 = prolog.query("find_similar_4('" + movie +"', M)")
  
  s5 = set()
  q5 = prolog.query("find_similar_5('" + movie +"', M)")

  return [handle_similarity(s1, q1, 1, show = False),
          handle_similarity(s2, q2, 2, show = False),
          handle_similarity(s3, q3, 3, show = False),
          handle_similarity(s4, q4, 4, show = False),
          handle_similarity(s5, q5, 5, show = False)]

find_similar("Pirates of the Caribbean: At Worlds End")
````

Η συνάρτηση `gridSearch()` αναζητά τη βέλτιστη παράμετρο για τη συνάρτηση `train_recommender()` η οποία εκπαιδεύει το σύστημα συστάσεων. Το όρισμα movieWindows είναι μια λίστα με τα μεγέθη παραθύρων που θα δοκιμάσουμε για την εκπαίδευση του συστήματος. Η συνάρτηση διατρέχει τα μεγέθη παραθύρου (naively) 
και εκπαιδεύει το σύστημα για το συγκεκριμένο μέγεθος παραθύρου. Για κάθε επανάληψη, υπολογίζεται η μέση βαθμολογία της σύστασης για 10 προβλέψεις.

Στις μεταβλητές `bestScore` και `bestParam` αποθηκεύουμε τα χαρακτηριστικά της βέλτιστης λύσης. Στο τέλος, εκτυπώνουμε το καλύτερο σκορ που βρήκαμε και το καλύτερο μέγεθος παραθύρου που οδήγησε σε αυτό.

````python
import tqdm

def gridSearch(movieWindows = [i for i in range(1, 53, 2)]):
    stats = {"precisions" : [], "recalls" : [], "f1s" : []}
    bestScore = 0
    bestParam = 1
    bestSim = 1
    max_length = len("--------------------------------------")
    for window in tqdm.tqdm(movieWindows):
        for similarity in range(1, 6):
            metrics = []
            for i in range (10):
                movie_score = train_recommender(train_ratings, rating_weights, score_weights, similarity, window, show=False)
                real, pred = predict_example(test_ratings, movie_score)
                metrics.append(get_metrics(real, pred))

            # print("+--------------------------------------+")
            # print(f"| Mean Score for {window}-movie window |")
            precisions = []
            recalls = []
            f1s = []
            for metric in metrics:
                precisions.append(metric["precision"])
                recalls.append(metric["recall"])
                f1s.append(metric["f1"])
                # print(f"| {metric}: {round(np.mean([m[metric] for m in metrics]))} |")
            
            stats["precisions"].append(round(np.mean(precisions), 3))
            stats["recalls"].append(round(np.mean(recalls), 3))
            stats["f1s"].append(round(np.mean(f1s), 3))
            print(f"""
                    +------------------------------------------------+
                    | Mean Score for {window}-movie window and similarity {similarity}{" "*(2 - len(str(window)))}|
                    | precision:  {str(round(np.mean(precisions), 3)).ljust(5, '0')}                              |
                    | recall:     {str(round(np.mean(recalls), 3)).ljust(5, '0')}                              |
                    | f1_score:   {str(round(np.mean(f1s), 3)).ljust(5, '0')}                              |
                    +------------------------------------------------+
                        """)
            
            if np.mean(f1s) > bestScore:
                bestScore = np.mean(f1s)
                bestParam = window
                bestSim = similarity
    print(f"Best Score: {round(bestScore, 3)}")
    print(f"Best Window: {round(bestParam, 3)}")
    print(f"Best Similarity: {round(bestSim, 3)}")

    return stats

windows = [i for i in range(1, 53, 2)]
stats = gridSearch(windows)
````
